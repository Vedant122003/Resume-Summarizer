```python
from sentence_transformers import SentenceTransformer


model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(['This is a test sentence.', 'This is a test collections.'])
print(embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(embeddings)
print(similarity)
```

```python
docs = [
    "Python is a programming language",
    "Cats are animals",
    "AI is the future"
]

modelNew = SentenceTransformer('all-MiniLM-L6-v2')
embeddingsNew = modelNew.encode(docs)
print(len(embeddingsNew))
print(len(embeddingsNew[0]))
```

```python
print(cosine_similarity(embeddingsNew))
```

```python
query = "What is Python?"
query_embedding = modelNew.encode([query])
similarity_scores = cosine_similarity(query_embedding, embeddingsNew)
print(similarity_scores)
```
